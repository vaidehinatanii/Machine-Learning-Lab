```python
import pandas as pd
import numpy as np

# Load dataset
url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"
df = pd.read_csv(url)

# Display first few rows
print("First 5 Records:")
print(df.head())

# Identify column types
print("\nDataset Info:")
print(df.info())

# Separate numerical and categorical columns
num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

print("\nNumerical Columns:", list(num_cols))
print("Categorical Columns:", list(cat_cols))

# Select numerical features for analysis
scores = df[['math score','reading score','writing score']]

# ---- Measures of Central Tendency ----
print("\nMean:")
print(scores.mean())

print("\nMedian:")
print(scores.median())

print("\nMode:")
print(scores.mode().iloc[0])

# ---- Measures of Dispersion ----
print("\nMinimum:")
print(scores.min())

print("\nMaximum:")
print(scores.max())

print("\nSum:")
print(scores.sum())

print("\nVariance:")
print(scores.var())

print("\nStandard Deviation:")
print(scores.std())

import pandas as pd

# Load dataset
url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"
df = pd.read_csv(url)

# Select numerical columns
scores = df[['math score', 'reading score', 'writing score']]

# Calculate quartiles
Q1 = scores.quantile(0.25)
Q2 = scores.quantile(0.50)   # Median
Q3 = scores.quantile(0.75)

# Display results
print("First Quartile (Q1):")
print(Q1)

print("\nSecond Quartile (Median / Q2):")
print(Q2)

print("\nThird Quartile (Q3):")
print(Q3)

# Correlation matrix
correlation = scores.corr()
print("Correlation Matrix:")
print(correlation)

# Covariance matrix
covariance = scores.cov()
print("\nCovariance Matrix:")
print(covariance)

import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
scores.hist(figsize=(8,5))
plt.suptitle("Score Distribution")
plt.show()

# Boxplots
plt.figure(figsize=(6,4))
sns.boxplot(data=scores)
plt.title("Boxplot of Student Scores")
plt.show()
```
